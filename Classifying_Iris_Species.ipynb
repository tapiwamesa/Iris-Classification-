***Our goal is to build a machine learning model that can learn from the measurements of these irises whose species is known, so that we can predict the species for a new iris***

*Since we have measurements for which we know the correct species of iris, this is a supervised learning problem. In this problem, we want to predict one of several options (the species of iris). The possible outputs (different species of irises) are called classes. Every iris in the dataset belongs to one of three classes, so this problem is a three-class classification problem.*

***Model Training***

```python
# we start by importing machine learning library sklearn and its dependencies

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Splitting data into test and train sets

X = data.drop(columns = ['target', 'target_name'])
y = data['target']
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
# viewing shape of the test set
X_test.shape, y_test.shape
```

```python
# viewing shape of the train set
X_train.shape, y_train.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Instantiating the KNN model

# k = 1
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# Instantiating the KNN model at k = 2

# k = 2
knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# Instantiating the KNN model at k = 3

# k = 3
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
# creating a loop that loops over changing values of k from 1 to 20

results = pd.DataFrame(columns = ["k", "Accuracy Score"])

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    results.loc[len(results)] = [k , accuracy]

results
```

```python
import matplotlib.pyplot as plt
```

```python
# plotting the accuracy score

plt.figure(figsize = (10, 5))
plt.plot(results["k"], results["Accuracy Score"], marker = "o", linestyle = "-")
plt.title("Accuracy over Evolution of k")
plt.legend("Accuracy Score")
plt.xticks(range(1, 20))
plt.show()
```

```python
# The accuracy score is maximized at k = 3, thus at 3 neighbors hence we adopt these as our
# optimum k for a generalized model for flower classification
```
